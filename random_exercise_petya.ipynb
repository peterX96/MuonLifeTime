```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
from scipy.optimize import curve_fit
from matplotlib.ticker import (MultipleLocator, AutoMinorLocator)

mpl.rcParams.update({'font.size':25})
```

```python
N = int(1e7)
M = int(1e4)
a, b = 2, 5

rand_arr = np.random.random(N)*(b-a)+a
```

```python
bin_contents = plt.hist(rand_arr, bins = M)[0]
```

```python
con_contents, contents = plt.hist(bin_contents, bins = 20)[:2]
contents = (contents[1:]+contents[:-1])/2
```

```python
def gaus(x, x0, sigma, A):
    return A*np.exp( -(x-x0)**2/2/sigma**2 )
```

```python
plt.scatter(contents, con_contents, label = 'bin contents')

popt, pcov = curve_fit(gaus, contents, con_contents, p0 = (1000, 50, 1500))

plt.plot(contents, gaus(contents, *popt), label = 'gaus fit')

plt.title(fr'$x_0$={round(popt[0])}, $\sigma$ = {round(popt[1], 1)}, ');
```
